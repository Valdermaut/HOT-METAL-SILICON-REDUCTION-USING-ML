```
import pingouin as pg
import pandas.util.testing as tm
```

```
df = pd.read_csv("base_data2.csv")
df.drop(['CLASS','DT','YEAR','HML_SI_CAT','HML_SI_BIN'],axis=1,inplace=True)
```

```
pg.corr(x=df['CORR_COKE_RATE'], y=df['CORR_FUEL_RATE'])
#Here the relation between CORR_COKE_RATE and CORR_FUEL_RATE is shown through value of r  = 0.809231~0.81
```

```
pg.corr(x=df['COLD_BLAST_O2_ENRICH'], y=df['O2_INJ_FLOW'])
#Here the relation between COLD_BLAST_O2_ENRICH and O2_INJ_FLOW is shown through value of r  = 0.965819~0.97
```

```
pg.corr(x=df['HML_CAO'], y=df['HML_CAO_SIO2'])
#Here the relation between HML_CAO and HML_CAO_SIO2 is shown through value of r  = 0.829125~0.83
```

```
df.corr().round(2)

#The correlation matrix for the following dataset
```

```
corrs = df.corr()
mask = np.zeros_like(corrs)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corrs, cmap='Spectral_r', mask=mask, square=True, vmin=-.4, vmax=.4)
plt.title('Correlation matrix')
```

```
corr = df.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df.columns),1.8)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```
